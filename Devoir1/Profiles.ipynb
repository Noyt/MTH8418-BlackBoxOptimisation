```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
PATH = '/home/noyt/Desktop/BA6/MTH8418/Devoir1/executions/executions'
data_no = pd.read_csv(PATH+'/GRIEWANK/4/NO_MODELS/res.txt', sep= '\s+', header = None)
data_quad = pd.read_csv(PATH+'/GRIEWANK/4/QUAD/res.txt', sep= '\s+', header = None)
data_tgp = pd.read_csv(PATH+'/GRIEWANK/4/TGP/res.txt', sep= '\s+', header = None)

#help(data_no.plot)
#data_no.plot(x = 'A', y='D')

sets = [data_no, data_quad, data_tgp]
for df in sets:
    df.columns = ['A', 'B', 'C', 'D']
    
global_max = max(data_no.A.max(), data_quad.A.max(), data_tgp.A.max())
global_min = min(data_no.A.min(), data_quad.A.min(), data_tgp.A.min())

uf=[]
for i in np.arange(3):
    idx = np.arange(global_min, global_max+1)
    sets[i] = sets[i].set_index('A')
    uf = sets[i]
    sets[i] = pd.DataFrame.from_dict(
               {i: x.values  for i, x in sets[i].groupby('A')['D']},
                orient='index').reindex(idx).ffill()

zog = pd.concat([sets[0], sets[1], sets[2]], axis=1)
zog.columns = ['NO_MODELS', 'QUAD', 'TGP']
#help(zog.plot)
zog.plot(logx=1, logy = 1)
```
